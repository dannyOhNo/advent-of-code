```python
# On each line, the calibration value can be found by combining the first digit 
# and the last digit (in that order) to form a single two-digit number.
# Consider your entire calibration document. What is the sum of all of the calibration values?
# If the sequence only has one number, it counts twice, e.g. trebu7chet = 77
# one2seven=22
```

```python
class Matcher(object):
    
    def __init__(self):
        self.number_arr = []
        
        # Pad int_map with int:int "1": 1
        for i in range(1, 10):
            self.int_map[str(i)] = i
        
    int_map = {
        "one": 1,
        "two": 2,
        "three": 3,
        "four": 4,
        "five": 5,
        "six": 6,
        "seven": 7,
        "eight": 8,
        "nine": 9
    }
    
    def get_int_mapping(self, s):
        """
            str:"one" -> int:1
            str:"1"   -> int:1
        """
        if s is None:
            return None
        
        for k,v in self.int_map.items():
            if k == s.lower():
                return v
        return None

    def load_file(self):
        """
        Load file return array with all test inputs
        """
        with open("inputs/1.txt", "r") as f:
            l = f.read()
        self.number_arr = l.split("\n")
        
    @staticmethod
    def is_int(x):
        """
        If it's an int, return true. Else try to convert to int, and return True. Else return False.
        """
        if isinstance(x, int):
            return True
        try:
            xi = int(x)
            return True
        except ValueError:
            return False
        
    
    def get_answer_from_map(self, results_map):
        """
            results_map: dictionary containing the value found and the string index position where it was found.
            -> Go through the map, find the lowest and highest index. The corresponding keys are the answer.
        """
        lv, hv = 10000, 0
        lvs, hvs = None, None
        
        for k,values in results_map.items():
            for v in values:
                if v > hv:
                    hv = v # "38" index
                    hvs = k # "one" search string
                if v < lv:
                    lv = v
                    lvs = k
                    
        # Convert to int, "eight" -> 8                   
        lvs_number = self.get_int_mapping(lvs) 
        hvs_number = self.get_int_mapping(hvs)
        
        # accomidate for single number, "7" -> "77"
        if not lvs or not hvs:
            r = lvs or hvs
            r = str(r) * 2
            return int(r)
        # Two numbers, "eight3" -> "83"
        else:
            res = int(str(lvs_number) + str(hvs_number))
            return res

    def get_number_indices(self, s):
        """
        Generate a dictionary containing the key ("two") and a value[list] containing the string index where the key was found.
        In the string "asd2asdff2", we would have { "2": [3, 9] }
        """
        results_map = {}
        for k, v in self.int_map.items(): # k = one, v = 1
            indices = [index for index in range(len(s)) if s.startswith(k, index)] # find all indexes of k (e.g. "one") in the string s
            if len(indices):
                if not isinstance(results_map.get(k), list):
                    results_map[k] = indices
        
        # Go through the map we just created, find the lowest and highest indexes.
        # Convert the corresponding key to an int, do a "string add" and return.
        answer = self.get_answer_from_map(results_map)
        
        return answer
        
                  
    def run(self):
        sum = 0
        
        self.load_file()

        for n in self.number_arr:
            answer = self.get_number_indices(n)
            sum += answer
            
        print(sum)
        
        
m = Matcher()
m.run()
```
